## Data Manipulation

```python
Gender = []
for item in data['GENDER']:
    if item == 'M':
       item = 1
    else:
        item = 0
    Gender.append(item)


    
```

```python
Lung_Cancer = []
for item in data['LUNG_CANCER']:
    if item == 'YES':
       item = 1
    else:
        item = 0
    Lung_Cancer.append(item)
```

```python
data['GENDER'] = Gender
```

```python
data['LUNG_CANCER']  = Lung_Cancer
```

## Spliting the Data

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop('LUNG_CANCER',axis=1)
```

```python
y= data['LUNG_CANCER']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

## Training the Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=200)
```

```python
rf.fit(x_train,y_train)
```

```python
pridiction = rf.predict(x_test)
```

## Evaluation

```python
from  sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pridiction))
print('/n')
print(classification_report(y_test,pridiction))
```
